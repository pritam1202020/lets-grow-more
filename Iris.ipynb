```python
#FEATURE ENGINEERING
#Breaking the dataset into 3 categories of species
data_setosa= data[data["Species"]=="Iris-setosa"]
data_versicolor= data[data["Species"]=="Iris-versicolor"]
data_virginia= data[data["Species"]=="Iris-virginia"]
```

```python
#outlier detection
upper_limit= data.SepalLengthCm.mean()+3*data.SepalLengthCm.std()
Lower_limit= data.SepalLengthCm.mean()-3*data.SepalLengthCm.std()
```

```python
data= data[(data["SepalLengthCm"] < upper_limit) & (data["SepalLengthCm"] > Lower_limit)]
```

```python
data
```

```python
x= data.iloc[:,1:5].values
x
```

```python
# choosing the optimum njumber of clusters for the training
from sklearn.cluster import KMeans 
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++',random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

The number of clusters will be "3"

```python
#Trainning the K-means cluster into the dataset
km=KMeans(n_clusters=3, init= "k-means++",random_state=42)
y_kmeans= km.fit_predict(x)
```

```python
print(y_kmeans)
```

```python
#visualize the cluster
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Iris-versicolor')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='Iris-setosa')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Iris-virginica')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,c='yellow',label='centroids')
plt.title("Cluster of speciece")
plt.legend()
plt.show()
```
